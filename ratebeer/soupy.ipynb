```python
from bs4 import BeautifulSoup
import requests
from HTMLParser import HTMLParser
import collections
import os
import json
import itertools
import nltk
from nltk.corpus import stopwords
```

```python
r = requests.get("http://www.ratebeer.com/beer/dogfish-head-90-minute-imperial-ipa/10569/")
html = r.text
```

```python
s = '/beer/dogfish-head-90-minute-imperial-ipa/10569/'
s.split('/')
```

```python
div_block_ind = html.find('COMMERCIAL DESCRIPTION')
substr = html[div_block_ind:]
start = substr.find('</small>')
substr = substr[start+len('</small>'):]
end = substr.find('</div>')
desc = substr[:end]
desc = desc.replace('<br>', '')
```

```python
parser = HTMLParser()
```

```python
desc = parser.unescape(desc)
desc = desc.encode('ascii')
```

```python
a = {'Stone': desc, 'Stone2': desc}
print len(a)
```

```python
for i,(u,v) in enumerate(a.iteritems()):
    print i, u, v
```

```python
with open('IPA.json', 'r') as f:
    ipa =  json.load(f)
```

```python
descList = [b.values()[0].split() for b in ipa]  # list of lists
descList = list(itertools.chain.from_iterable(descList)) #list
descList = map(unicode.lower, descList)
print descList
#wc = collections.Counter(chain)
```

```python
s = set(stopwords.words('english'))
#print s
keywords = filter(lambda w: not w in s, descList) # filter out common words
wc = collections.Counter(keywords)
wc.most_common(n=40)
```
