```python
import collections
import json
import itertools
import nltk
from nltk.corpus import stopwords
from glob import glob
import numpy as np
import pandas as pd
from matplotlib import pyplot as py
```

```python
files = glob('*.json')
n = len(files)
beerDict = {}
for fn in files:
    style = fn.split('.')[-2]
    with open(fn, 'r') as f:
        beerDict[style] =  json.load(f)
        
```

```python
for sty, beer in beerDict.iteritems():
    descList = [b.values()[0].split() for b in beer]  # list of lists
    descList = list(itertools.chain.from_iterable(descList)) #list
    descList = map(unicode.lower, descList)
    print descList
```

```python
beerDict2 = {sty: {} for sty in beerDict.keys()}
for sty in beerDict2.keys():
    for beer in beerDict[sty]:
        x = beer.items()
        beerDict2[sty][x[0][0]] = x[0][1]
```

```python
dd = collections.defaultdict(list)
#d = {dd}

#dstyle = collections.defaultdict(list)
#dvalues = collections.defaultdict(list)
#dindex = collections.defaultdict(list)

for (sty, d) in beerDict2.iteritems():
    for beer, desc in d.iteritems():
        dd['Style']+=[sty]
        dd['Description']+=[desc]
        dd['Beer']+=[beer]
#dd
df = pd.DataFrame(dd, dtypes=[])
#df
```

```python
def splitDesc(desc):
    desc = desc.lower()
    desc = desc.replace('.', ' ')
    desc = desc.replace('&', ' ')
    desc = desc.replace('-', ' ')
    desc = desc.replace('<br>', '')
    desc = desc.replace('<br />', '')
    desc = desc.split()
    s = set(stopwords.words('english'))
    keywords = filter(lambda w: not w in s, desc)
    return keywords
```

```python
df['Keywords'] = df['Description'].apply(splitDesc)
```

```python
lagers = df[df['Style'] == 'amber_lager']
ipas = df[df['Style'] == 'IPA']
pales = df[df['Style'] == 'american_pale_ale']

countList = []
labels = ['Lagers', 'IPAs', 'Pales']
totalCollection = collections.Counter()

for beer in [lagers, ipas, pales]:
    x = list(itertools.chain.from_iterable(beer['Keywords'].values)) 
    wc = collections.Counter(x)
    countList.append(wc.most_common(n=50))
    totalCollection += wc

    
#ipas = list(itertools.chain.from_iterable(ipas['Keywords'].values))
#wc = collections.Counter(ipas)
#wc.most_common(n=50)


pd.DataFrame(countList[1]).plot()
py.show()
```
